## Merging all the datasets based on the country variable

```python
## Since the year 2017 was not appearing on the columns names, I am adding the year suffixes to each column before merging

# Function to add year suffix to column names (excluding 'Country')
def add_year_suffix(df, year):
    df = df.rename(columns=lambda x: f"{x}_{year}" if x != "Country" else x)
    return df

# Add year suffix to columns
df_2015 = add_year_suffix(df_2015, 2015)
df_2016 = add_year_suffix(df_2016, 2016)
df_2017 = add_year_suffix(df_2017, 2017)
df_2018 = add_year_suffix(df_2018, 2018)
df_2019 = add_year_suffix(df_2019, 2019)
```

```python
# Merge datasets step by step
all_years_merge = df_2015.merge(df_2016, on='Country', suffixes=('_2015', '_2016'))
all_years_merge = all_years_merge.merge(df_2017, on='Country', suffixes=('', '_2017'))
all_years_merge = all_years_merge.merge(df_2018, on='Country', suffixes=('', '_2018'))
all_years_merge = all_years_merge.merge(df_2019, on='Country', suffixes=('', '_2019'))
```

```python
# Display all the columns
pd.set_option('display.max_columns', None)
```

```python
all_years_merge
```

```python
# Deleting the years columns
columns_to_drop = [col for col in all_years_merge.columns if 'Year' in col]
all_years_merge = all_years_merge.drop(columns=columns_to_drop)
```

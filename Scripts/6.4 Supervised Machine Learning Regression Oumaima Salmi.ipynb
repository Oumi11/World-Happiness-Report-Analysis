## Importing librairies, path and datasets

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Defining the path 
path = r'C:\Users\salmi\Desktop\achievement 6'
```

```python
# Importing the dataframe 
all_years_merge_2 = pd.read_csv(os.path.join(path, 'Data sets', 'Prepared Data Happiness', 'all_years_merge_2_update_ex_3.csv'), index_col = False)

new_happiness_data = pd.read_csv(os.path.join(path, 'Data sets', 'Prepared Data Happiness', 'new_happiness_data_only_avg_factors_update_ex_3.csv'), index_col = False)
```

## Exploratory analysis

```python
# Create a scatterplot using matplotlib for another look at how the chosen variables plot against each other.

plot_2_eco_happiness = new_happiness_data.plot(x = 'Averaged Economy', y='Averaged Happiness Score',style='o') # The style option creates a scatterplot; without it, we only have lines.
plt.title('Economy vs Happiness Score')  
plt.xlabel('Economy (GDP  per Capita)')  
plt.ylabel('Happiness Score')  
plt.show()
```

```python
# Reshape the variables into NumPy arrays and put them into separate objects.

X = new_happiness_data['Averaged Economy'].values.reshape(-1,1)
y = new_happiness_data['Averaged Happiness Score'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Split data into a training set and a test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

#### Hypothesis statement

1. If the economy of a country is strong, then the happiness score is higher.

## Regression analysis

#### Economy vs Happiness Score

```python
# Create a regression object.

regression = LinearRegression()
```

```python
# Fit the regression object onto the training set.

regression.fit(X_train, y_train)
```

```python
# Predict the values of y using X.

y_predicted = regression.predict(X_test)
```

```python
# Create a plot that shows the regression line from the model on the test set.

plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Economy vs Happiness Score (Test set)')
plot_test.xlabel('Economy (GDP per Capita')
plot_test.ylabel('Happiness Score')
plot_test.show()
```

```python
# Create objects that contain the model summary statistics.

rmse = mean_squared_error(y_test, y_predicted) # This is the mean squared error
r2 = r2_score(y_test, y_predicted) # This is the R2 score. 
```

```python
# Print the model summary statistics. This is where you evaluate the performance of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```

## Answer to the question

The positive slope in both the test set and training set indicate a positive relationship between the economy and the happiness score. A stronger economy is associated with higher happiness level.

The Rscore on the training set performed a bit better (0,68 for 0,63 in the test) and indicates that the relatively close values indicate that the moddel generalizes well to the new data. The model explains then a significant portion of the variance in happiness scores (68%). There is however 32% of the variance that is unexplained by the model. This makes sense as the happiness score is not only build on the economy factors. 

To conclude, the relatively positive performance metrics suggest that the model is reliable. However, the small amount of data here (only 140 countries or data points) could not be reliable to implement that on other new data. Indeed, this could lead to inacurracy and biased results.
